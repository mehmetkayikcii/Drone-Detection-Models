```python
import cv2
import torch
import numpy as np
from pathlib import Path
import logging
import warnings

# Deprecation uyarılarını filtrele
warnings.filterwarnings(
    "ignore",
    message=r".*torch\.cuda\.amp\.autocast.*deprecated.*"
)

# Logging ayarları
date_fmt = '%Y-%m-%d %H:%M:%S'
logging.basicConfig(format='%(asctime)s %(levelname)s: %(message)s', datefmt=date_fmt, level=logging.INFO)

# Yol tanımları
image_dir = Path("/content/yolov5/Drones-3/test/images")
label_dir = Path("/content/yolov5/Drones-3/test/labels")
output_dir = Path("/content/test-sonuclari-yolov5")
output_dir.mkdir(parents=True, exist_ok=True)

# Klasör kontrolü
if not image_dir.exists():
    logging.error(f"Görüntü klasörü bulunamadı: {image_dir}")
    raise FileNotFoundError(f"Görüntü klasörü bulunamadı: {image_dir}")
if not label_dir.exists():
    logging.warning(f"Etiket klasörü bulunamadı, işlemler etiketsiz devam edecek: {label_dir}")

# Modeli yükle
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
model = torch.hub.load(
    'ultralytics/yolov5',
    'custom',
    path='/content/drive/MyDrive/best-drones-oddbp.pt',
    device=device
)
logging.info(f"Model {device} üzerinde yüklendi")

# Görselleri işle
for image_path in image_dir.glob("*.jpg"):
    img = cv2.imread(str(image_path))
    if img is None:
        logging.warning(f"Görüntü okunamadı: {image_path.name}")
        continue
    h, w = img.shape[:2]

    # Gerçek etiketleri oku ve çiz (yeşil)
    label_path = label_dir / (image_path.stem + ".txt")
    if not label_dir.exists() or not label_path.exists():
        logging.warning(f"Etiket dosyası bulunamadı: {label_path.name}")
    else:
        with open(label_path, "r") as f:
            for line in f:
                cls, x, y, bw, bh = map(float, line.strip().split())
                x1 = int((x - bw / 2) * w)
                y1 = int((y - bh / 2) * h)
                x2 = int((x + bw / 2) * w)
                y2 = int((y + bh / 2) * h)
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(
                    img,
                    f"gercek etiket: {int(cls)}",
                    (x1, max(y1 - 10, 0)),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.5,
                    (0, 255, 0),
                    1,
                    cv2.LINE_AA
                )

    # Model tahminleri (kırmızı)
    results = model(str(image_path))
    for *xyxy, conf, cls in results.xyxy[0]:
        x1, y1, x2, y2 = map(int, xyxy)
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
        cls_id = int(cls.item())
        label = model.names[cls_id] if hasattr(model, 'names') else str(cls_id)
        cv2.putText(
            img,
            f"tahmin-yolov5: {label} {conf:.2f}",
            (x1, min(y2 + 15, h - 1)),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5,
            (0, 0, 255),
            1,
            cv2.LINE_AA
        )

    # Sonuçları kaydet (ek ismiyle)
    out_name = image_path.stem + "_cmp.jpg"
    cv2.imwrite(str(output_dir / out_name), img)
    logging.info(f"Kaydedildi: {out_name}")
```

```python
import os
import glob
import torch
import json
from torchvision.transforms import functional as F
from torchvision.models.detection import retinanet_resnet50_fpn
from torchvision.models.detection.anchor_utils import AnchorGenerator
from PIL import Image

# --- Yol ayarları ---
checkpoint_path = "/content/drive/MyDrive/retinanet_drone.pth"
test_images_dir = "/content/yolov5/Drones-3/test/images"
output_json     = "/content/test_predictions.json"

# --- Cihaz seçimi ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- Checkpoint yükle ---
checkpoint = torch.load(checkpoint_path, map_location=device)
state_dict = checkpoint.get("model_state_dict", checkpoint)

# --- Anchor generator tanımı (eğitimde kullanılanla uyumlu) ---
anchor_sizes = ((32, 64, 128),) * 5
anchor_ratios = ((0.5, 1.0, 2.0),) * 5
anchor_generator = AnchorGenerator(sizes=anchor_sizes, aspect_ratios=anchor_ratios)

# --- num_classes belirle ---
out_channels = state_dict["head.classification_head.cls_logits.bias"].shape[0]
anchors_per_loc = len(anchor_sizes[0]) * len(anchor_ratios[0])
num_classes = out_channels // anchors_per_loc

# --- Model oluştur ve ağırlıkları yükle ---
model = retinanet_resnet50_fpn(
    num_classes=num_classes,
    anchor_generator=anchor_generator
)
model.load_state_dict(state_dict)
model.to(device).eval()

# --- Test verisi üzerinde inference ---
results = []
image_paths = sorted(glob.glob(os.path.join(test_images_dir, "*.jpg")))
for img_id, img_path in enumerate(image_paths):
    # Görseli yükle ve tensöre çevir
    img = Image.open(img_path).convert("RGB")
    tensor = F.to_tensor(img).to(device).unsqueeze(0)

    # Tahmin
    with torch.no_grad():
        output = model(tensor)[0]

    # Kutular, sınıflar ve skorları al
    boxes  = output['boxes'].cpu().tolist()
    labels = output['labels'].cpu().tolist()
    scores = output['scores'].cpu().tolist()

    # Deadline, confidence eşiği uygula (ör: 0.5)
    for box, label, score in zip(boxes, labels, scores):
        if score < 0.5:
            continue
        x1, y1, x2, y2 = box
        w = x2 - x1
        h = y2 - y1
        results.append({
            "image_id": os.path.basename(img_path),
            "category_id": label,
            "bbox": [x1, y1, w, h],
            "score": score
        })

# JSON dosyasına kaydet
with open(output_json, 'w') as f:
    json.dump(results, f, indent=2)
print(f"✅ Test tahminleri kaydedildi: {output_json}")
```

```python
import os
import json
from collections import defaultdict
from PIL import Image, ImageDraw, ImageFont

def visualize_saved_predictions(
    predictions_json: str,
    images_dir: str,
    labels_dir: str,
    output_dir: str,
    score_threshold: float = 0.5
):
    # Dizinleri oluştur
    os.makedirs(output_dir, exist_ok=True)

    # Font seçimi
    try:
        font = ImageFont.truetype(
            "/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 18
        )
    except IOError:
        font = ImageFont.load_default()

    # Tahminleri yükle ve image_id bazında grupla
    with open(predictions_json, 'r') as f:
        results = json.load(f)
    preds_by_image = defaultdict(list)
    for r in results:
        if r.get("score", 0) >= score_threshold:
            preds_by_image[r["image_id"]].append(r)

    # Her bir görüntü için
    for image_id, preds in preds_by_image.items():
        img_path = os.path.join(images_dir, image_id)
        if not os.path.exists(img_path):
            print(f"Uyarı: Görsel bulunamadı, atlanıyor: {img_path}")
            continue

        # Görüntüyü yükle
        image = Image.open(img_path).convert("RGB")
        draw = ImageDraw.Draw(image)
        img_w, img_h = image.size

        # Tahminleri çiz (kırmızı)
        for pred in preds:
            x1, y1, w, h = pred["bbox"]
            x2, y2 = x1 + w, y1 + h
            label = f"tahmin: {pred['category_id']} {pred['score']:.2f}"
            draw.rectangle([x1, y1, x2, y2], outline="red", width=2)
            draw.text((x1, y1 - 15), label, fill="red", font=font)

        # Gerçek etiketleri çiz (yeşil)
        label_path = os.path.join(
            labels_dir,
            os.path.splitext(image_id)[0] + ".txt"
        )
        if os.path.exists(label_path):
            with open(label_path, 'r') as lf:
                for line in lf:
                    cls, cx, cy, bw, bh = map(float, line.strip().split())
                    tx1 = (cx - bw/2) * img_w
                    ty1 = (cy - bh/2) * img_h
                    tx2 = (cx + bw/2) * img_w
                    ty2 = (cy + bh/2) * img_h
                    gt_label = f"gerçek etiket: {int(cls)}"
                    draw.rectangle([tx1, ty1, tx2, ty2], outline="green", width=2)
                    draw.text((tx1, ty1 - 15), gt_label, fill="green", font=font)

        # Kaydet
        out_path = os.path.join(output_dir, image_id)
        image.save(out_path)
        print(f"Kaydedildi: {out_path}")


if __name__ == "__main__":
    visualize_saved_predictions(
        predictions_json="/content/test_predictions.json",
        images_dir="/content/yolov5/Drones-3/test/images",
        labels_dir="/content/yolov5/Drones-3/test/labels",
        output_dir="/content/drive/MyDrive/visualized_results",
        score_threshold=0.5
    )
```

```python
import os
import json
from PIL import Image, ImageDraw, ImageFont

# --- Ayarlar ---
image_dir = "/content/yolov5/Drones-3/test/images"
gt_label_dir = "/content/yolov5/Drones-3/test/labels"
output_dir = "/content/combined_results"
os.makedirs(output_dir, exist_ok=True)

# Her model için JSON dosyaları ve stil ayarları
predictions_info = {
    "yolov5": {
        "json": "/content/test-sonuclari-yolov5/predictions.json",  # ya da test-sonuclari-yolov5 içindeki .json
        "color": "red",
        "name": "YOLOv5"
    },
    "fastercnn": {
        "json": "/content/test-sonuclari-fastercnn/predictions.json",
        "color": "blue",
        "name": "FasterR-CNN"
    },
    "retinanet": {
        "json": "/content/retinanet-tahmin2/predictions_retinanet.json",
        "color": "orange",
        "name": "RetinaNet"
    }
}

# JSON’dan tahminleri yükle ve image_id bazında grupla
def load_predictions(path):
    with open(path, 'r') as f:
        data = json.load(f)
    grouped = {}
    for p in data:
        img = p['image_id']
        grouped.setdefault(img, []).append(p)
    return grouped

preds = {k: load_predictions(v['json']) for k, v in predictions_info.items()}

# Font
try:
    font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 16)
except IOError:
    font = ImageFont.load_default()

# Görselleştirme döngüsü
for img_name in os.listdir(image_dir):
    if not img_name.lower().endswith(('.jpg', '.png')):
        continue
    img_path = os.path.join(image_dir, img_name)
    image = Image.open(img_path).convert('RGB')
    draw = ImageDraw.Draw(image)
    w_img, h_img = image.size

    # 1) Ground Truth (yeşil)
    gt_path = os.path.join(gt_label_dir, os.path.splitext(img_name)[0] + '.txt')
    if os.path.exists(gt_path):
        with open(gt_path) as f:
            for line in f:
                cls, cx, cy, bw, bh = map(float, line.strip().split())
                x1 = (cx - bw/2) * w_img
                y1 = (cy - bh/2) * h_img
                x2 = (cx + bw/2) * w_img
                y2 = (cy + bh/2) * h_img
                draw.rectangle([x1, y1, x2, y2], outline='green', width=2)
                draw.text((x1, y1-12), f"GT: {int(cls)}", fill='green', font=font)

    # 2) Her modelin tahminleri
    for key, info in predictions_info.items():
        model_preds = preds[key].get(img_name, [])
        for p in model_preds:
            bx, by, bw, bh = p['bbox']
            x1, y1 = bx, by
            x2, y2 = bx + bw, by + bh
            color = info['color']
            label = info['name']
            score = p.get('score', None)
            draw.rectangle([x1, y1, x2, y2], outline=color, width=2)
            txt = f"{label}" + (f" {score:.2f}" if score is not None else "")
            draw.text((x1, y1-12), txt, fill=color, font=font)

    # Kaydet
    out_path = os.path.join(output_dir, img_name)
    image.save(out_path)
    print(f"Kaydedildi: {out_path}")
```

```python
import os
import json
import torch
import torchvision
from torchvision.transforms import functional as F
from torchvision.models.detection import retinanet_resnet50_fpn
from torchvision.models.detection.anchor_utils import AnchorGenerator
from PIL import Image, ImageDraw, ImageFont
from pathlib import Path

# ----------------------------
# Birleştirilmiş Görselleştirme
# ----------------------------

# Yol tanımları
image_dir        = Path("/content/yolov5/Drones-3/test/images")
gt_label_dir     = Path("/content/yolov5/Drones-3/test/labels")
output_dir       = Path("/content/combined_results")
output_dir.mkdir(parents=True, exist_ok=True)

# JSON dosyaları (Faster R-CNN tahminleri)
faster_json = "/content/drive/MyDrive/fasterrcnn_test_predictions.json"
with open(faster_json, 'r') as f:
    faster_preds = json.load(f)
# Grupla
faster_by_img = {}
for p in faster_preds:
    faster_by_img.setdefault(p['image_id'], []).append(p)

# YOLOv5 modeli
device       = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
yolo_model   = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/drive/MyDrive/best-drones-oddbp.pt', device=device)
yolo_model.eval()

# RetinaNet modeli
checkpoint   = torch.load('/content/retinanet_drone2.pth', map_location=device)
state_dict   = checkpoint.get('model_state_dict', checkpoint)
# AnchorGenerator tanımı
anchor_sizes  = ((32,64,128),)*5
anchor_ratios = ((0.5,1.0,2.0),)*5
ag = AnchorGenerator(sizes=anchor_sizes, aspect_ratios=anchor_ratios)
# num_classes
out_ch        = state_dict['head.classification_head.cls_logits.bias'].shape[0]
anchors_loc   = len(anchor_sizes[0]) * len(anchor_ratios[0])
num_classes   = out_ch // anchors_loc
# Model yükle
retina_model = retinanet_resnet50_fpn(num_classes=num_classes, anchor_generator=ag)
retina_model.load_state_dict(state_dict)
retina_model.to(device).eval()

# Font
try:
    font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 18)
except IOError:
    font = ImageFont.load_default()

# Görselleştirme
for img_path in sorted(image_dir.glob("*.jpg")):
    img_name = img_path.name
    img      = Image.open(img_path).convert('RGB')
    draw     = ImageDraw.Draw(img)
    w, h     = img.size

    # 1) Ground truth (yeşil)
    gt_path = gt_label_dir / f"{img_path.stem}.txt"
    if gt_path.exists():
        with open(gt_path, 'r') as f:
            for line in f:
                cls, cx, cy, bw, bh = map(float, line.strip().split())
                x1 = (cx - bw/2) * w;  y1 = (cy - bh/2) * h
                x2 = (cx + bw/2) * w;  y2 = (cy + bh/2) * h
                draw.rectangle([x1,y1,x2,y2], outline='green', width=2)
                draw.text((x1, y1-12), f"gerçek etiket: {int(cls)}", fill='green', font=font)

    # 2) YOLOv5 tahminleri (kırmızı)
    results = yolo_model(str(img_path))
    for *xyxy, conf, cls_id in results.xyxy[0]:
        x1, y1, x2, y2 = map(int, xyxy)
        label = yolo_model.names[int(cls_id)]
        draw.rectangle([x1,y1,x2,y2], outline='red', width=2)
        draw.text((x1, y2+4), f"tahmin-yolov5: {label} {conf:.2f}", fill='red', font=font)

    # 3) Faster R-CNN tahminleri (mavi)
    for p in faster_by_img.get(img_name, []):
        bx, by, bw, bh = p['bbox']
        x1, y1 = bx, by;  x2, y2 = bx + bw, by + bh
        score   = p.get('score', 0)
        cat     = p.get('category_id', None)
        draw.rectangle([x1,y1,x2,y2], outline='blue', width=2)
        draw.text((x1, y2+4), f"tahmin-fastercnn: {cat} {score:.2f}", fill='blue', font=font)

    # 4) RetinaNet tahminleri (turuncu)
    tensor  = F.to_tensor(img).to(device).unsqueeze(0)
    with torch.no_grad():
        out = retina_model(tensor)[0]
    for box, score, lbl in zip(out['boxes'].cpu(), out['scores'].cpu(), out['labels'].cpu()):
        if score < 0.5: continue
        x1,y1,x2,y2 = map(int, box.tolist())
        draw.rectangle([x1,y1,x2,y2], outline='orange', width=2)
        draw.text((x1, y1-12), f"tahmin : retinanet {score:.2f}", fill='orange', font=font)

    # Kaydet
    save_path = output_dir / img_name
    img.save(save_path)
    print(f"Kaydedildi: {save_path}")
```

```python
import os
import json
import torch
from PIL import Image, ImageDraw, ImageFont
from pathlib import Path
import torchvision
from torchvision.transforms.functional import to_tensor

# ----------------------------
# Birleştirilmiş Görselleştirme (Legend ile)
# ----------------------------

# Yol tanımları
data_root       = Path("/content")
image_dir       = data_root / "yolov5/Drones-3/test/images"
gt_label_dir    = data_root / "yolov5/Drones-3/test/labels"
faster_json     = data_root / "drive/MyDrive/fasterrcnn_test_predictions.json"
retina_json     = data_root / "drive/MyDrive/predictions_retinanet3.json"
output_dir      = data_root / "sonuclar"
output_dir.mkdir(parents=True, exist_ok=True)

# Grup yükleme fonksiyonu
def load_grouped(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)
    grouped = {}
    for p in data:
        grouped.setdefault(p['image_id'], []).append(p)
    return grouped

# 1) Faster R-CNN tahminlerini JSON’dan yükle ve grupla
faster_by_img = load_grouped(faster_json)
# 2) RetinaNet tahminlerini JSON’dan yükle ve grupla
retina_by_img = load_grouped(retina_json)

# 3) YOLOv5 modeli yükle
device     = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
yolo_model = torch.hub.load(
    'ultralytics/yolov5', 'custom',
    path=str(data_root / 'drive/MyDrive/best-drones-oddbp.pt'),
    device=device
)
yolo_model.eval()

# Font ayarı
try:
    font = ImageFont.truetype(
        "/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 18
    )
except IOError:
    font = ImageFont.load_default()

# Legend öğeleri
legend_items = [
    ("gercek etiket: (Yesil)", "green"),
    ("YOLOv5: (Kirmizi)", "red"),
    ("Faster R-CNN: (Mavi)", "blue"),
    ("RetinaNet: (Turuncu)", "orange"),
]

# Görselleştirme döngüsü
for img_path in sorted(image_dir.glob("*.jpg")):
    img_name = img_path.name
    img      = Image.open(img_path).convert('RGB')
    draw     = ImageDraw.Draw(img)
    w, h     = img.size

    # a) Ground Truth (yeşil)
    gt_path = gt_label_dir / f"{img_path.stem}.txt"
    if gt_path.exists():
        for line in open(gt_path, 'r'):
            cls, cx, cy, bw, bh = map(float, line.strip().split())
            x1 = (cx - bw/2) * w;  y1 = (cy - bh/2) * h
            x2 = (cx + bw/2) * w;  y2 = (cy + bh/2) * h
            draw.rectangle([x1,y1,x2,y2], outline='green', width=2)

    # b) YOLOv5 tahminleri (kırmızı)
    results = yolo_model(str(img_path))
    for *xyxy, conf, cls_id in results.xyxy[0]:
        x1, y1, x2, y2 = map(int, xyxy)
        draw.rectangle([x1,y1,x2,y2], outline='red', width=2)

    # c) Faster R-CNN tahminleri (mavi)
    for p in faster_by_img.get(img_name, []):
        bx, by, bw, bh = p['bbox']
        x1, y1 = bx, by;  x2, y2 = bx + bw, by + bh
        draw.rectangle([x1,y1,x2,y2], outline='blue', width=2)

    # d) RetinaNet tahminleri (turuncu)
    for p in retina_by_img.get(img_name, []):
        bx, by, bw, bh = p['bbox']
        x1, y1 = bx, by;  x2, y2 = bx + bw, by + bh
        draw.rectangle([x1,y1,x2,y2], outline='orange', width=2)

    # Legend çizimi
    margin = 20
    box_size = 20
    x0 = w - margin - 200
    y0 = margin
    for label_text, color in legend_items:
        draw.rectangle([x0, y0, x0+box_size, y0+box_size], fill=color, outline=color, width=2)
        draw.text((x0+box_size+10, y0), label_text, fill=color, font=font)
        y0 += box_size + 10

    # Kaydet
    img.save(output_dir / img_name)
    print(f"Kaydedildi: {output_dir/img_name}")
```
